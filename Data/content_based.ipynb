```python
new_book_tags = pd.read_csv("content_based.csv")
```

```python
new_book_tags["tag_concat"] = new_book_tags.groupby(["goodreads_book_id", "original_title"]).transform(lambda x : "|".join(x))
```

```python
new_book_tags
```

```python
new_book_tags.drop(columns=["tag_id", "count", "Unnamed: 0", "tag_name"], inplace=True)
```

```python
new_book_tags = new_book_tags.iloc[::5,:]
```

```python
new_book_tags.to_csv("new_book_tags.csv")
```

```python
new_book_tags_onehot = new_book_tags.join(new_book_tags['tag_concat'].str.get_dummies("|"))
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cos_sim = cosine_similarity(new_book_tags_onehot.iloc[:,5:])
```

```python
harry_id = 3
harry_top10 = np.argsort(cos_sim[harry_id])
harry_top10
```

```python
harry_top10 = harry_top10[-10:][::-1]
```
